```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from numpy.random import rand, randn, randint
```

```python
#Creating sample Dataframe and Pandas series

df = pd.DataFrame(randn(1000), index  = pd.date_range('2019-01-01', periods = 1000), columns = ['value'])
ts = pd.Series(randn(1000), index  = pd.date_range('2019-01-01', periods = 1000))

df.head()
```

```python
#cumsum() This is cummulative sum function used to get the cumulative sum of all the values in a column

df['value'] = df['value'].cumsum()
df.head()
```

```python
ts.head()
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
ts.plot(figsize = (25,5)) #We have given the size of the figure plot
```

```python
fig = plt.figure(figsize = (8,8)) #This create empty area (canvas) for the figure
plt.plot(ts)
fig.savefig('abc.png')  #You can give any extension
```
