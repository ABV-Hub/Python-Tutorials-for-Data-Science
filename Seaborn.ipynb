# Categorical Plots

```python
#Categorical Graph

sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, jitter = False)

#If you need a stright line plot
```

```python
#To find the deviation from the mean in categorical graph

sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm')
```

```python
#Box plot

sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box', hue = 'sex')
```

```python
#Load Dataset: DIAMONDS

diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
#Bixen plot (Same like Box plot. Used for beautification purpose)

sns.catplot(x = 'color', y = 'price', data = diamonds, kind = 'boxen')
```

```python
#Voline Plot

sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin', hue = 'sex')

#These plot how us the deviation of the data from the mean value
```

```python
#Combining violin and swarm plot together

g = sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin') #We created violin plot and saved it in a variable 'g'

sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm', ax = g.ax, color = 'k')
#We creeated a swarm plot and gave its axis 'ax' to be the same as that of 'g'
```

```python
#Load Titanic dataset

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#Count Plot

sns.catplot(x = 'deck', data = titanic, kind = 'count', hue = 'class')

#Since we are counting the number, we have to prvide only one axis either x or y
```
